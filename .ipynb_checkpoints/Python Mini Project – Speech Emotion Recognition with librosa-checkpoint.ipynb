# Python Mini Project – Speech Emotion Recognition with librosa

Speech Emotion Recognition, abbreviated as SER, is the act of attempting to recognize human emotion and affective states from speech. This is capitalizing on the fact that voice often reflects underlying emotion through tone and pitch. This is also the phenomenon that animals like dogs and horses employ to be able to understand human emotion.
SER is tough because emotions are subjective and annotating audio is challenging.

Librosa, a Python package for analysing audio and music, will be used in this project. It standardises interfaces and names, provides a flatter package layout, backwards compatibility, modular functions, and readable code. To create a model with an MLPClassifier, we'll work with the libraries librosa, soundfile, and sklearn, among others. This system will be able to identify emotions in audio files. After loading the data and extracting its features, the dataset will be divided into training and testing sets. The model will then be trained after initialising an MLPClassifier. We'll calculate our mode's accuracy lastly. The RAVDESS (Ryerson Audio-Visual Database of Emotional Speech and Song) dataset can be found at https://drive.google.com/file/d/1wWsrN2Ep7x6lWqOXfr4rpKGYrJhWc8z7/view. In this dataset, 7356 files were evaluated on emotional validity, intensity, and sincerity by 247 people ten times. 24 actors contributed to the 24.8GB total dataset.

1. Make the neccessary imports

```python
import librosa
import soundfile
import os, glob, pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

2. To define function to extract the mfcc, chroma, and mel features from a sound file

Mel Frequency Cepstral Coefficient (mfcc) : This gauges a sound's transient power spectrum.

Chroma : Refers to the 12 various pitch classes

mel: Frequency of the Mel Spectrogram.
    
With soundfile, open the audio file. SoundFile is using with-as so that when we're finished, it closes itself. Call it X and read from it. Obtain the sample rate as well. Get the Short-Time Fourier Transform of X if chroma is True.

Let result be a numpy array that is empty. Get the mean value by calling the corresponding function from librosa.feature for each of the three features, if it exists (for example, librosa.feature.mfcc for mfcc). With result and the feature value as arguments, use hstack() from numpy and store the result. hstack() horizontally stacks arrays in order (in a columnar fashion). Return the outcome next.

```python
def extract_feature(file_name, mfcc, chroma, mel):
    with soundfile.SoundFile(file_name) as sound_file:
        X = sound_file.read(dtype="float32")
        sample_rate=sound_file.samplerate
        if chroma:
            stft=np.abs(librosa.stft(X))
        result=np.array([])
        if mfcc:
            mfccs=np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)
            result=np.hstack((result, mfccs))
        if chroma:
            chroma=np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
            result=np.hstack((result, chroma))
        if mel:
            mel=np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
            result=np.hstack((result, mel))
    return result
```

3. Let us create a dictionary to store the numbers and the RAVDESS dataset's accessible emotions, and a list to hold the desired emotions (calm, happy, terrified, and disgust).

```python

emotions={
  '01':'neutral',
  '02':'calm',
  '03':'happy',
  '04':'sad',
  '05':'angry',
  '06':'fearful',
  '07':'disgust',
  '08':'surprised'
}

observed_emotions=['calm', 'happy', 'fearful', 'disgust']
```

4. Let's use the function load data(), which accepts the relative size of the test set as an argument, to load the data now. We'll use the glob() method from the glob module to get all of the pathnames for the sound files in our dataset because x and y are empty lists. Get the file's basename and emotion for each of these paths by splitting the name around the letter "-" and extracting the third value. This number is converted into an emotion using our emotions dictionary, and then our function checks to see if that emotion is on our list of observed emotions; if not, it moves on to the next file. It calls extract feature and stores the result in the variable "feature." The characteristic and feeling are then added to x and y, respectively. Thus, list X has the features, while list Y contains the emotions. With these, the test size, and a random state value, we call the function train test split and return it.

```python
def load_data(test_size=0.2):
    x,y=[],[]
    for file in glob.glob("data\\Actor_*\\*.wav"):
        file_name=os.path.basename(file)
        emotion=emotions[file_name.split("-")[2]]
        if emotion not in observed_emotions:
            continue
        feature=extract_feature(file, mfcc=True, chroma=True, mel=True)
        x.append(feature)
        y.append(emotion)
    return train_test_split(np.array(x), y, test_size=test_size, random_state=9)
```

5. The dataset should now be divided into training and testing sets. Utilizing the load data method, let's maintain the test set at 25% of everything.

```python
x_train,x_test,y_train,y_test=load_data(test_size=0.25)
```

6. Examine how the training and testing datasets are structured.

```python
print((x_train.shape[0], x_test.shape[0]))
```

7. And get the number of features extracted.

```python
print(f'Features extracted: {x_train.shape[1]}')
```

8. Let's now start up an MLPClassifier. This multi-layer perceptron classifier uses stochastic gradient descent or LBFGS to optimise the log-loss function. The MLPClassifier uses an internal neural network for classification, unlike SVM or Naive Bayes. This ANN model is feedforward.

```python
np.random.seed(42)
model=MLPClassifier(alpha=0.01, batch_size=256, epsilon=1e-08, hidden_layer_sizes=(300,), learning_rate='adaptive', max_iter=500)
```

9. Fit/train the model

```python
model.fit(x_train,y_train)
```

10. Let’s predict the values for the test set. This gives us y_pred (the predicted emotions for the features in the test set).

```python
y_pred=model.predict(x_test)
```

11. We'll use the accuracy score() method from sklearn that we imported to determine the accuracy of our model. The accuracy will then be rounded to two decimal places and printed.

```python
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)

print("Accuracy: {:.2f}%".format(accuracy*100))
```

12. We will use confusion_matrix function to determine the confusion matrix of our model.

```python
cm = confusion_matrix(y_true = y_test, y_pred = y_pred)
print(cm)
```

13. Plot confusion matrix in a beatiful manner

```python
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax, fmt = 'g'); #annot=True to annotate cells
# labels, title and ticks
ax.set_xlabel('Predicted', fontsize=20)
ax.xaxis.set_label_position('top') 
ax.xaxis.set_ticklabels(['calm', 'happy', 'fearful', 'disgust'], fontsize = 15)
ax.xaxis.tick_top()

ax.set_ylabel('True', fontsize=20)
ax.yaxis.set_ticklabels(['calm', 'happy', 'fearful', 'disgust'], fontsize = 15)
plt.show()
```
